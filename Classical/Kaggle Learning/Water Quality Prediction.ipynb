```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
```

```python
# load data in
water_data_path = r'C:\Users\Paul\OneDrive - University of Exeter\Machine Learning Summer Code\Classical\water_potability.csv'
water_data = pd.read_csv(water_data_path)

# set X and y
water_data.dropna(axis=0, subset=['Potability'], inplace=True)
y = water_data.Potability
water_data.drop(['Potability'], axis=1, inplace=True)
X = water_data.select_dtypes(exclude=['object'])

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=0, train_size=0.8, test_size=0.2)

# data has some null values so using imputation to assign average values

my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

imputed_X_train.columns = X_train.columns
imputed_X_train.columns = X_valid.columns

final_X_train = imputed_X_train
final_X_valid = imputed_X_valid

pota_model = RandomForestRegressor(n_estimators=100, random_state=0)
pota_model.fit(final_X_train, y_train)
pota_preds = pota_model.predict(final_X_valid)
print(mean_absolute_error(y_valid, pota_preds))

```
